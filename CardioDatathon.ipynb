```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**STARTING EDA**

```python
Data = pd.read_csv('/content/cardio_train.csv',delimiter = ';')
```

```python
Data.head()
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data.isnull().sum()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(Data.corr(), annot=True, cmap='coolwarm')
```

```python
Data['height'] = Data['height']/100
```

```python
Data['bmi']= Data['weight']/ (Data['height'] ** 2)
```

```python
Data.head()
```

```python
Data.drop(['height','weight'], axis=1, inplace=True)
```

```python
Data['bp_ratio']= Data['ap_hi']/ Data['ap_lo']
```

```python
Data.drop(['ap_hi','ap_lo'], axis=1, inplace=True)
```

```python
Data.head()
```

```python
Data = Data[['id','age','gender','bmi','cholesterol','gluc','bp_ratio','alco','smoke','active','cardio']]
```

```python
Data.head()
```

```python
Data.drop('id', axis=1, inplace=True)
```

```python
Data.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(Data.corr(), annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(Data['bmi'], kde=True)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cardio', y='age', data=Data)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='smoke', y='cardio', data=Data)
```
